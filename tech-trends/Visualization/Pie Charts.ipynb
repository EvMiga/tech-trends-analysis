### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
db=df[['DatabaseWantToWorkWith']].dropna()
db.loc[:,'DatabaseWantToWorkWith']=db['DatabaseWantToWorkWith'].str.split(';')
db=db.explode('DatabaseWantToWorkWith')
freq=db['DatabaseWantToWorkWith'].value_counts()
top5 = freq.head(5)

plt.figure(figsize=(12,8))
plt.pie(
    x=top5.values,
    labels=top5.index,
    autopct='%1.1f%%',
    shadow='True'
)
plt.title('Top 5 Data Base')
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
dev=df[['DevType']].dropna()
freq=dev['DevType'].value_counts()
top5=freq.head(5)

plt.figure(figsize=(12,8))
plt.pie(
    x=top5.values,
    labels=top5.index,
    autopct='%1.1f%%',
    shadow=True
)
plt.title('Top 5 dev role')
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here
os=df[['OpSysProfessional use']].dropna()
count=os['OpSysProfessional use'].value_counts()
top10=count.head(10)

plt.figure(figsize=(12,8))
plt.pie(
    x=top10.values,
    labels=top10.index,
    autopct='%1.1f%%',
    shadow=True
)
plt.title('Top 10 the best OS')
plt.show()
```

### Task 3: Analyzing and Interpreting Composition


In this task, you will create additional pie charts to analyze specific aspects of the survey data. Use `pandas` and `matplotlib` to complete each task and interpret the findings.



##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
##Write your code here
la=df[['LanguageAdmired']].dropna()
la.loc[:,'LanguageAdmired']=la['LanguageAdmired'].str.split(';')
la=la.explode('LanguageAdmired')
best=la['LanguageAdmired'].value_counts()
bl=best.head(5)

plt.figure(figsize=(12,8))
plt.pie(
    x=bl.values,
    labels=bl.index,
    autopct='%1.1f%%',
    shadow=True
)
plt.title('Top 5 admired languages')
plt.show()
```

##### 3.2 Pie Chart of Tools Used for AI Development


Using the `AIToolCurrently` Using column, create a pie chart to visualize the top 5 tools developers are currently using for AI development.



```python
##Write your code here
ai=df[['AIToolCurrently Using']].dropna()
ai.loc[:,'AIToolCurrently Using']=ai['AIToolCurrently Using'].str.split(';')
ai=ai.explode('AIToolCurrently Using')
ai_t=ai['AIToolCurrently Using'].value_counts()
ai_tools=ai_t.head(5)

plt.figure(figsize=(12,8))
plt.pie(
    x=ai_tools.values,
    labels=ai_tools.index,
    autopct='%1.1f%%',
    shadow=True
)
plt.title('Top 5 tools developers are currently using for AI development')
plt.show()
```

##### 3.3 Pie Chart for Preferred Web Frameworks


The `WebframeWantToWorkWith` column includes web frameworks that respondents are interested in working with. Visualize the top 5 frameworks in a pie chart.



```python
##Write your code here
fr=df[['WebframeWantToWorkWith']].dropna()
fr.loc[:,'WebframeWantToWorkWith']=fr['WebframeWantToWorkWith'].str.split(';')
fr=fr.explode('WebframeWantToWorkWith')
c=fr['WebframeWantToWorkWith'].value_counts()
f5=c.head(5)

plt.figure(figsize=(12,8))
plt.pie(
    x=f5.values,
    labels=f5.index,
    autopct='%1.1f%%',
    shadow=True
)
plt.title('Web frameworks that respondents are interested in working with')
plt.show()
```

##### 3.4 Pie Chart for Most Desired Embedded Technologies


Using the `EmbeddedWantToWorkWith` column, create a pie chart to show the top 5 most desired embedded technologies that respondents wish to work with.



```python
##Write your code here
it=df[['EmbeddedWantToWorkWith']].dropna()
it.loc[:,'EmbeddedWantToWorkWith']=it['EmbeddedWantToWorkWith'].str.split(';')
it=it.explode('EmbeddedWantToWorkWith')
co=it['EmbeddedWantToWorkWith'].value_counts()
it5=co.head(5)

plt.figure(figsize=(12,8))
plt.pie(
    x=it5.values,
    labels=it5.index,
    autopct='%1.1f%%',
    shadow=True,
    pctdistance=0.75,
    labeldistance=1.05,
    wedgeprops={'edgecolor': 'white', 'linewidth': 3}
)
plt.title('Top 5 most desired embedded technologies that respondents wish to work with')
plt.show()
```
