Collect the number of vacancies by programming language with API: C C# C++ Java JavaScript Python Scala Oracle SQL Server MySQL Server PostgreSQL MongoDB

```python
import requests
from openpyxl import Workbook
api_url = "http://127.0.0.1:5000/data"
def number_of_jobs_S(Skill):
    response = requests.get(api_url, params={'Key Skills':Skill})
    data = response.json()
    number_of_jobs = len(data)
    return number_of_jobs
Skills = ["C",
          "C#",
          "C++",
          "Java",
          "JavaScript",
          "Python",
          "Scala",
          "Oracle",
          "SQL Server",
          "MySQL Server",
          "PostgreSQL",
          "MongoDB"
         ]
wb = Workbook()
ws = wb.active
ws.title = "Job by Skills"
ws.append(['Skills', 'number_of_jobs'])

for S in Skills:
    count = number_of_jobs_S(S)
    ws.append([S,count])
    
wb.save('skills_jobs.xlsx')
    
    
```
